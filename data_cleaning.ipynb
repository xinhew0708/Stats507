## Question 3

```python
# a.
# Read the four files
df1 = pd.read_sas('DEMO_G.XPT')
df2 = pd.read_sas('DEMO_H.XPT')
df3 = pd.read_sas('DEMO_I.XPT')
df4 = pd.read_sas('DEMO_J.XPT')
```

```python
# Keep required columns
index = ['SEQN', 'RIDAGEYR', 'RIDRETH3', 'DMDEDUC2', 'DMDMARTL',
        'RIDSTATR', 'SDMVPSU', 'SDMVSTRA', 'WTMEC2YR', 'WTINT2YR']
df1 = df1.loc[:, index]
df1['year'] = '2011-2012'
df2 = df2.loc[:, index]
df2['year'] = '2013-2014'
df3 = df3.loc[:, index]
df3['year'] = '2015-2016'
df4 = df4.loc[:, index]
df4['year'] = '2017-2018'
```

```python
# Concatenate
df = pd.concat([df1, df2, df3, df4])

# Rename columns
names = dict(zip(index, ['id', 'age', 'race', 'education', 'marital', 
                         'status', 'psu_variance', 'stratum_variance', 
                         'mec_exam_weight', 'interview_weight']))
df = df.rename(columns=names)
```

```python
# Convert types
int_index = ['id', 'age']
df[int_index] = df[int_index].astype(int)

c_index = ['race', 'education', 'marital', 'status', 'year']
df[c_index] = df[c_index].astype('category')
```

```python
# Replace categorical values
d = {1: "Mexican American",
     2: "Other Hispanic",
     3: "Non-Hispanic White",
     4: "Non-Hispanic Black", 
     6: "Non-Hispanic Asian",
     7: "Other"}
df[['race']] = df[['race']].replace(d)

d = {1: "married", 2: "widowed", 
     3: "divorced", 4: "separated",
     5: "never married", 6: "living with partner", 
     77: "refused", 99: "don't know"}
df[['marital']] = df[['marital']].replace(d)

d = {1: 'interview', 2: 'both'}
df[['status']] = df[['status']].replace(d)
df.head()
```

```python
# Report case numbers
print("There are", df[['id']].count()[0], 
      "cases in the demographic datasets.")
```

```python
# Save the dataset
df.to_pickle('df_demographic.pickle')
```

```python
# b.
# Read the four files
dfo1 = pd.read_sas('OHXDEN_G.XPT')
dfo2 = pd.read_sas('OHXDEN_H.XPT')
dfo3 = pd.read_sas('OHXDEN_I.XPT')
dfo4 = pd.read_sas('OHXDEN_J.XPT')
```

```python
# Keep required columns
index = ['SEQN', 'OHDDESTS'] + \
        [name for name in list(dfo1.columns) 
         if 'OHX' and 'TC' in name and 'CTC' not in name] + \
        [name for name in list(dfo1.columns) if 'OHX' and 'CTC' in name]
a = len([name for name in list(dfo1.columns) 
         if 'OHX' and 'TC' in name and 'CTC' not in name])
b = len([name for name in list(dfo1.columns) if 'OHX' and 'CTC' in name])

dfo1 = dfo1.loc[:, index]
dfo1['year'] = '2011-2012'
dfo2 = dfo2.loc[:, index]
dfo2['year'] = '2013-2014'
dfo3 = dfo3.loc[:, index]
dfo3['year'] = '2015-2016'
dfo4 = dfo4.loc[:, index]
dfo4['year'] = '2017-2018'
```

```python
# Concatenate
dfo = pd.concat([dfo1, dfo2, dfo3, dfo4])

# Rename columns
names = ['id', 'dentition_status'] + \
        ['tooth_count' + str(num) for num in range(a)] + \
        ['coronal_cavity' + str(num) for num in range(b)]

dfo = dfo.rename(columns=dict(zip(index, names)))
```

```python
# Convert types
int_index = ['id']
dfo[int_index] = dfo[int_index].astype(int)

c_index = ['dentition_status', 'year'] + \
          ['tooth_count' + str(num) for num in range(a)] + \
          ['coronal_cavity' + str(num) for num in range(b)]
dfo[c_index] = dfo[c_index].astype('category')
```

```python
# Replace categorical values
d = {1: "complete", 2: "partial", 3: "not done"}
dfo[['dentition_status']] = dfo[['dentition_status']].replace(d)
dfo.head()
```

```python
# Report case numbers
print("There are", dfo[['id']].count()[0], 
      "cases in the oral health and dentition datasets.")
```

```python
# Save dataset
dfo.to_pickle('df_oral.pickle')
```
